```python
%%time
if 1 == 0:
    for kkk in range(len(file_list)):
    #for kkk in range(len(file_list)-1,len(file_list)):
        print("Start: " + str(kkk))

        v0 = file_list[kkk].split("/")[-1].split("_")
        file_id = v0[0] + "_" + v0[1]
        df = pd.read_excel(file_list[kkk])
        out = []
        for i in range(4,len(df.iloc[3,:])):
            tmp = df.iloc[3,i] + "-" + df.iloc[4,i][1:] 
            out.append(tmp)
            #print(tmp)
        df_col = pd.DataFrame(out)
        df_col.columns = ["node"]

        out = []
        for i in range(5,len(df.iloc[:,3])):
            tmp = df.iloc[i,2] + "-" + df.iloc[i,3][1:] 
            out.append(tmp)
        df_row = pd.DataFrame(out)
        df_row.columns = ["node"]

        nodes = set(df_row["node"])
        nodes = list(nodes.intersection(set(df_col["node"])))
        nodes.sort()

        nodes_dict = dict()
        index2nodes = dict()
        for i in range(len(nodes)):
            nodes_dict.update({nodes[i]:i})
            index2nodes.update({i:nodes[i]})
            
        df_index2nodes = pd.DataFrame(index2nodes.items())
        df_index2nodes.columns = ["index","name"]


        
        df_mat = df.iloc[5:,4:].copy()
        edges = []
        for i in range(len(df_row)):
            if i % 1000 == 0:
                print(i)
            for j in range(len(df_col)):
                if i != j:
                    if df_col["node"].iloc[j] in nodes_dict:
                        if df_row["node"].iloc[i] in nodes_dict:
                            if df_mat.iloc[i,j] > 0:
                                send_index = nodes_dict[df_row["node"].iloc[i]]
                                rece_index = nodes_dict[df_col["node"].iloc[j]]
                                edges.append([send_index,rece_index,df_mat.iloc[i,j]])
        df_edges = pd.DataFrame(edges)
        
        input_file = "/home/rh/Arbeitsraum/bitbucket/share/tempgraph/open_data/WIOD2016/" + file_id + "_edgelist.csv"
        df_edges.to_csv(input_file,index=False,header=None)   

        input_file_2 = "/home/rh/Arbeitsraum/bitbucket/share/tempgraph/open_data/WIOD2016/" + file_id + "_index2nodes.csv"
        df_index2nodes.to_csv(input_file_2,index=False) 
        
        
        
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import digraphwave as digw
import data_utils as datautils

file_list_edge = glob.glob("/home/rh/Arbeitsraum/bitbucket/share/tempgraph/open_data/WIOD2016/*_edgelist.csv")
file_list_edge.sort()


file_list_index = glob.glob("/home/rh/Arbeitsraum/bitbucket/share/tempgraph/open_data/WIOD2016/*_index2nodes.csv")
file_list_index.sort()

input_file = file_list_edge[-1]
input_file_2 = file_list_index[-1]
df_edges = pd.read_csv(input_file,header=None)
df_index2nodes = pd.read_csv(input_file_2)

#input_file = "bank_monthmax_agg.csv"
filetype = "csv"  # ["csv", "tsv", "edgelist"]

print(input_file)

list_countries = []
for i in range(len(df_index2nodes)):
    v1 = df_index2nodes["name"].iloc[i].split("-")
    list_countries.append(v1[0])
    
list_countries = list(set(list_countries))
print(list_countries)
print(len(df_edges))
print(len(df_index2nodes))
```

# Set parameters

```python
%%time
is_weighted = True
is_directed = True
adj = datautils.read_to_spmat(input_file, filetype=filetype, is_weighted=is_weighted,
                                  directed=is_directed)
```

```python
%%time
radius = 5
kemb = 128
batch_size = 2000
is_arctan_log_transform = False  # DigraphwaveHyperparametersのデフォルト
n_flag = False  # DigraphwaveHyperparametersのデフォルト
t_flag = True  # DigraphwaveHyperparametersのデフォルト
a_flag = True  # DigraphwaveHyperparametersのデフォルト
order = 10  # DigraphwaveHyperparametersのデフォルト

param = digw.DigraphwaveHyperparameters.create(num_nodes=adj.shape[0], num_edges=adj.nnz,
                                                   R=radius, k_emb=kemb, batch_size=batch_size,
                                                   arctan_log_transform=is_arctan_log_transform,
                                                   n_flag=n_flag, 
                                                   t_flag=t_flag,
                                                   a_flag=a_flag,
                                                   order=order
                                                   )
```

# Calculate Embeddings

```python
%%time
num_gpu = 1
device_ids = list(range(num_gpu)) if num_gpu > 0 else None
embeddings = digw.digraphwave(adj=adj, param=param, device_ids=device_ids)

tsne = TSNE(n_components=2, random_state = 0, perplexity = 30, n_iter = 1000)
X_embedded = tsne.fit_transform(embeddings)
```
